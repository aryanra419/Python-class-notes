```python
What is data visualization?
it is the process of representing data in a graphical or
visual format such as graph, charts and maps so that people
can easily understand pattens, trends and insight

why visualization is important?
    *make data easy to understand
    *help in quick decision making
    *identifiers patterns and trends
    *detect outlier or error
    impoves communication of insight
```

```python
Univariate analysis:
it mean analyzing only one variable at a time.
    to check the quality of the data.
purpose:
* understand distribution
* find patterns
* detect outliers
ex: histogram, Box plot, KDE plot, Violin plot

Ex: analyze salary distribution in a company
->are most employes low paid or high paid


bivariate analysis:
it means analying relationship between two
variables.
purpose:
* understand relationship
* identify correlation
* compare variable
ex: scatter plot, bar plot, line plot

ex: analyze advertising spends Vs sales
Does increase ads increase sales?

multivariate analysis:
it means analyzing relationship between more tham two variables.

purpose:
*understand complex relationships
*identify the patterns across multipkle variables
*perform deep data analysis(EDA)
EDA: exploratory data analysis

ex: Pair plot, heatmap, facet grid, bubble plot
ex:analyze sales based on region,time and
product category
-> which combination gives highest profit?
```

```python
# matplotlib and seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.get_dataset_names()
```

```python
histogram:
* show distribution of data
* helps in detect skewness and spread  # one sided data is known as skewness
```

```python
plt.figure()
plt.hist(df["total_bill"])
plt.title("histogram")
plt.xlabel("Total Bill")
plt.ylabel("frequency")
plt.show()
```

```python
df["total_bill"].max()
```

```python
df["total_bill"].min()
```

```python
df["total_bill"].mean()
```

```python
df["total_bill"].median()
```

```python
KDE plot:
* smooth verion of histogram
* better for understanding data density
```

```python
plt.figure()
sns.kdeplot(df["total_bill"],fill=True)
plt.title("KDE plot")
plt.xlabel("Total Bill")
plt.ylabel("frequency")
plt.show()
```

```python
box plot:
* identifies outliers
* shows median,quartiles, spread
```

```python
plt.figure()
sns.boxplot(x=df["total_bill"])
plt.title("Box Plot")
plt.xlabel("Total Bill")
plt.show()
```

```python
Violin plot
* combines box plot + distribution
* shows data shapes clearly
```

```python
plt.figure()
sns.violinplot(x=df["total_bill"])
plt.title("Violin Plot")
plt.show()
```

```python
Bar Plot
* compare values across categories
* easy to understand differences
```

```python
plt.figure()
sns.barplot(x="day" ,y ="total_bill" , data=df)
plt.title("Bar Plot")
plt.show()
```

```python
plt.figure()
sns.barplot(x="total_bill" ,y ="day" , data=df,orient="h")
plt.title("Bar Plot")
plt.show()
```

```python
pie chart
* shows proportion/percentage
* good for quick insights
note: use only for small num of categories
```

```python
plt.figure()
df["day"].value_counts().plot.pie(autopct="%1.2f%%")
plt.title("Pie Chart")
plt.show()
```

```python
line plot
* shows trend over time
* helps to identify increase,decrease,seasonality
ex: sales growth month-wise
```

```python
plt.plot(df["total_bill"])
plt.title("Line Plot")
plt.show()
```

```python
Count plot
* show frequency of categories
* useful for categorial analysis
ex: no of customer per city
```

```python
sns.countplot(x="day",data=df)
plt.title("count plot")
plt.show()
```

```python
scatter plot:
* show relationship between two numerical values
* helps detect correlation and pattern
ex: Advertsing vs Sales
```

```python
sns.scatterplot(x="total_bill",y="tip",data=df)
plt.title("scatter plot")
plt.show()
```

```python
bubble sort:
* add tthird variable(size)
* shows more information in one chat

ex: sales Vs Profit with market size
```

```python
plt.scatter(df["total_bill"],df["tip"],s=df["size"]*30)
plt.title("scatter plot")
plt.show()
```

```python
Pair Plot
* shows relationship between all variables
* useful in EDA(exploratory data analysis)
 ex: feature relationshil in machine learning
```

```python
sns.pairplot(df)
plt.show()
```

```python
Heatmap:
* shows correlation between variables
* helps in feature selection
ex: identify highly correlated feature
```

```python
corr = df.corr(numeric_only=True)
corr.round(2)
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```
