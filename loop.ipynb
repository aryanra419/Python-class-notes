```python
num=121
temp =num

rev=0
while(num>0):
    rem =num%10
    rev=(rev*10)+rem
    num=num//10
if(rev==temp):
    print("its palindrome")
else:
    print("not palindrome")
```

```python
## '''to check the number is prime or not        divisible by 2,3,4,5,6,7,8       '''

n=int(input("Enter a number"))
count=0
i=2
while(i<n):
    if(n%i==0):
        count=count+1
    else:
        count=count
    i= i+1
    
if(count==0):
    print("its prime number")
else:
    print("not prime")
```
