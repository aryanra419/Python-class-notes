```python
What is data visualization?
it is the process of representing data in a graphical or
visual format such as graph, charts and maps so that people
can easily understand pattens, trends and insight

why visualization is important?
    *make data easy to understand
    *help in quick decision making
    *identifiers patterns and trends
    *detect outlier or error
    impoves communication of insight
```

```python
Univariate analysis:
it mean analyzing only one variable at a time.
    to check the quality of the data.
purpose:
* understand distribution
* find patterns
* detect outliers
ex: histogram, Box plot, KDE plot, Violin plot

Ex: analyze salary distribution in a company
->are most employes low paid or high paid


bivariate analysis:
it means analying relationship between two
variables.
purpose:
* understand relationship
* identify correlation
* compare variable
ex: scatter plot, bar plot, line plot

ex: analyze advertising spends Vs sales
Does increase ads increase sales?

multivariate analysis:
it means analyzing relationship between more tham two variables.

purpose:
*understand complex relationships
*identify the patterns across multipkle variables
*perform deep data analysis(EDA)
EDA: exploratory data analysis

ex: Pair plot, heatmap, facet grid, bubble plot
ex:analyze sales based on region,time and
product category
-> which combination gives highest profit?
```

```python
# matplotlib and seaborn

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.get_dataset_names()
```

```python
histogram:
* show distribution of data
* helps in detect skewness and spread  # one sided data is known as skewness
```

```python
plt.figure()
plt.hist(df["total_bill"])
plt.title("histogram")
plt.xlabel("Total Bill")
plt.ylabel("frequency")
plt.show()
```

```python
df["total_bill"].max()
```

```python
df["total_bill"].min()
```

```python
df["total_bill"].mean()
```

```python
df["total_bill"].median()
```

```python
KDE plot:
* smooth verion of histogram
* better for understanding data density
```

```python
plt.figure()
sns.kdeplot(df["total_bill"],fill=True)
plt.title("KDE plot")
plt.xlabel("Total Bill")
plt.ylabel("frequency")
plt.show()
```

```python
box plot:
* identifies outliers
* shows median,quartiles, spread
```

```python
plt.figure()
sns.boxplot(x=df["total_bill"])
plt.title("Box Plot")
plt.xlabel("Total Bill")
plt.show()
```
