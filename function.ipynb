```python
def prime2(st,sp):
    for i in range(st,sp+1,1):
        for j in range(2,i,1):
            if(i%j==0):
                break
        else:
            print(i,end=" ")

prime2(2,20)
```

```python
def prime3():
    pl = []
    for i in range(2,21,1):
        for j in range(2,i,1):
            if(i%j==0):
                break
        else:
            pl.append(i)
    return pl
prime3()
```

```python
def prime4(st,sp):
    pl =[]
    for i in range(st,sp+1,1):
        for j in range(2,i,1):
            if(i%j==0):
                break
        else:
            pl.append(i)
    return pl

prime4(2,21)
    
```

```python
recursive function:
a recursive fuction is a fucnction that calls itself until a condition is met

important parts
1- base condition - stops recursion
2-recursive call - function calling itself
```

```python
#sum of n numbers
def sum_n(n):
    if n==0:
        return 0
    else:
        return n+sum_n(n-1)

sum_n(10)
```

```python
#factorial number

def fact(n):
    if n==1:
        return 1
    else:
        return n*fact(n-1)
fact(5)
```
