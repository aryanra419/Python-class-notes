```python
exception handling
we use Try-Except

try:
    block of code
except:
    error msg
```

```python
try:
    a=int(input("Enter a"))
    b=int(input("Enter b"))
    c=a/b
    print(c)
except:
    print("invalid input")
```

```python
##Catching specific error
try:
    num = int(input("enter number"))
    print(10/num)
except ZeroDivisionError:
    print("you cannot divide by zero")
except ValueError:
    print("please enter a valid number")
```

```python
# try-except-else-finally

try:
    num = int(input("enter number"))
    print(10/num)
except ZeroDivisionError:
    print("you cannot divide by zero")
except ValueError:
    print("please enter a valid number")
else:
    print("No error occured")
finally:
    print("Execution completed")
```
