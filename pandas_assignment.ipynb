```python
1. Calculate Total Sales Amount for each transaction using a new column


2. Find total sales by each Region.


3. Display all transactions where Discount is greater than 5%.


4. Group data by Category and find the average UnitPrice for each.


5. Find the top 2 SalesPersons with the highest total sales amount.


6. Filter records for only Electronics category where sales are above ₹50,000.


7. Sort the dataset by Date and SalesPerson in ascending order.


8. Find the number of transactions for each CustomerType (Regular vs New).


9. Get the most frequently sold product in each region.


10. Pivot Table: Create a pivot table showing total sales by Region and Category.
```

```python
df["TotalSales"] = df["Quantity"] * df["UnitPrice"] * (1 - df["Discount"]/100)
df
```

```python
region_sales = df.groupby("Region")["TotalSales"].sum()
print(region_sales)
```

```python
discount = df[df["Discount"] > 5]
discount
```

```python
avg_price = df.groupby("Category")["UnitPrice"].mean()
print(avg_price)
```

```python
top_salespersons = df.groupby("SalesPerson")["TotalSales"].sum().sort_values(ascending=False).head(2)
print(top_salespersons)
```

```python
sales = df[(df["Category"] == "Electronics") & (df["TotalSales"] > 50000)]
sales
```

```python
sort = df.sort_values(by=["Date", "SalesPerson"])
sort
```

```python
customer_count = df["CustomerType"].value_counts()
print(customer_count)
```

```python
product = df.groupby("Region")["Product"].agg(lambda x: x.value_counts().idxmax())
print(product)
```

```python
pivot_table = pd.pivot_table(df, values="TotalSales", index="Region", columns="Category", aggfunc="sum")
pivot_table
```
